## The libraries to import zipfile --change the path accordingly

```python
import pandas as pd
import zipfile
import os

# Path to the zip file
zip_file_path = r"C:\Users\ameen\Desktop\JrDataEngg\FinalProject\Dataset.zip"

# Directory to extract the files
extract_dir = r"C:\Users\ameen\Desktop\JrDataEngg\FinalProject\extracted_files"

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the extracted files
extracted_files = os.listdir(extract_dir)
print("Extracted files:")
print(extracted_files)
```

## Create Dictionary  to store the multiple Dataframe from each csv file

```python

# Directory containing the extracted CSV files
extract_dir = r"C:\Users\ameen\Desktop\JrDataEngg\FinalProject\extracted_files"

# List of CSV files
csv_files = [
    "cleaned_customers.csv"
    "cleaned_geolocation.csv"
    "cleaned_orders_items.csv"
    "cleaned_orders_payments.csv"
    "cleaned_order_reviews.csv"
    "cleaned_orders.csv"
    "cleaned_product_category_name_translation.csv"
    "cleaned_products.csv"
    "cleaned_sellers.csv"
]

# Dictionary to store dataframes
dfs = {}

# Create dataframes for each CSV file
for csv_file in csv_files:
    file_path = os.path.join(extract_dir, csv_file)
    df_name = csv_file.replace(".csv", "").replace(" ", "_")  # Use file name as dataframe name
    dfs[df_name] = pd.read_csv(file_path)

# Example usage:
# Access the dataframe for 'olist_customers_dataset.csv' like this:
# df_customers = dfs['olist_customers_dataset']
```

## To display each dataframe:

```python
# Display all dataframes
for df_name, df in dfs.items():
    print("DataFrame:", df_name)
    display(df.head())  # Display the first few rows of each dataframe
    print("\n")
```
